#DFS(depth first search) for graph and visualization step by step

```python

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)

    print(start)

    for next_node in graph[start] - visited:
        dfs(graph, next_node, visited)
    return visited

graph = {
    'A': {'B', 'C'},
    'B': {'A', 'D', 'E'},
    'C': {'A', 'F'},
    'D': {'B'},
    'E': {'B', 'F'},
    'F': {'C', 'E'}
}


dfs(graph, 'A')
```

```python
import matplotlib.pyplot as plt
import networkx as nx


G = nx.Graph(graph)

plt.figure(figsize=(8, 5))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=700, font_size=15, font_weight='bold', edge_color='gray')
plt.title("Graph Visualization")
plt.show()
```

```python
def draw_graph_static(G, visited_nodes, pos):
    plt.figure(figsize=(8, 5))
    nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700)
    nx.draw_networkx_nodes(G, pos, nodelist=visited_nodes, node_color='red', node_size=700)
    nx.draw_networkx_edges(G, pos, edge_color='gray')

    nx.draw_networkx_labels(G, pos, font_size=15, font_weight='bold')

    plt.title("Graph Visualization")
    plt.show()

def dfs_visualization_static(graph, start, pos, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    draw_graph_static(G, visited, pos)

    for next_node in graph[start] - visited:
        dfs_visualization_static(graph, next_node, pos, visited)

    return visited


pos = nx.spring_layout(G)
dfs_visualization_static(graph, 'A', pos)
```
